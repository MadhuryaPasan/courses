### Define variables

```R
name <- readline("What's your name? ")
greeting <- paste("Hello,", name)# concatenate strings
print(greeting)
#output -> [1] "Hello, user"
```
*`combining different strings  in to one string is called "string concatenation"`*
eg -> "Hello,"      "World" ->  "Hello, World"

### Documentation

in the console type `?paste` to get the documentation about `paste` funtion. the documentation will display on the help pannel

```R
name <- readline("What's your name? ")
greeting <- paste("Hello,", name, sep="_")
print(greeting)
#output -> [1] "Hello,_user"
```

*if you do not want to add sep there is an another way. this will concatenate string without seperators*
```R
name <- readline("What's your name? ")
greeting <- paste("Hello,", name) 
print(greeting)

#output -> [1] "Hello,user"
```

### Storage Modes
`Character, Double, Integer, ...`

`Coercion -> changing storage mode of an object (as.character | as.double | as.integer)`

```R
mario <- readline("Enter votes for Mario: ")
peach <- readline("Enter votes for Peach: ")
bowser <- readline("Enter votes for Bowser: ")

mario <- as.integer(mario)
peach <- as.integer(peach)
bowser <- as.integer(bowser)

total <- mario + peach + bowser
print(paste("Total votes:", total))
```

or

```R
mario <- as.integer(readline("Enter votes for Mario: "))
peach <- as.integer(readline("Enter votes for Peach: "))
bowser <- as.integer(readline("Enter votes for Bowser: "))

total <- mario + peach + bowser
print(paste("Total votes:", total))
```


### sum()

### Acsses Rows and Columns

* `variable[row,column]` -> one cell
* `variable[,1]` -> all the values of first column
* `variable[1,]` -> all the values of first row
* `variable$column` -> get the column
* `variable$column[1]` -> get a value inside a column

these returns a vector

### Sum of a column

or
```R
sum(votes$poll)
```
