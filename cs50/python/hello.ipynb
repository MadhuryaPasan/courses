## round

round(number[,ndigits]) 
`[ ] -> means that, inside of this is optional items`

```python
x = float(input("What's x?"))
y = float(input("What's y?"))

z = round(x+y)

print(z)
```

```python
x = float(input("What's x?"))
y = float(input("What's y?"))

z = round(x / y , 2) # round up to last 2 decimal points

print(z)
```

```python
# using format string (f-string)

x = float(input("What's x?"))
y = float(input("What's y?"))

z = x / y

print(f"{z:.2f}")
```

## formating number output 

`1000 -> 1,000`

```python
x = float(input("What's x?"))
y = float(input("What's y?"))

z = round(x+y)

print(f"{z:,}")
```

## def
~define~

```python
def hello():
    print("hello")

name = input("What's your name? ")
hello()
print(name)
```

```python
# include print function in def

def hello(to):
    print("hello,", to)

name = input("What's your name? ")
hello(name)
```

```python
# default value
def hello(to = "World"):
    print("hello,", to)

name = input("What's your name? ")
hello(name)
hello()
```

## def main():

```python
def main():
    name = input("What's your name? ")
    hello(name)


def hello(to = "World"):
    print("hello,", to)


main()

# now the hello function can be included below the main.
```
