## match - case


```python
name = input("What's your name? ")

match name:
    case "Harry":
        print("Gryffindor")
    case "Hermione":
        print("Gryffindor")
    case "Ron":
        print("Gryffindor")
    case "Draco":
        print("Slythering")
    case _:
        print("Who?")
```

```python
# short code

name = input("What's your name?")

match name:
    case "Harry" | "Hermione" | "Ron":
        print("Gryffindor")
    case "Draco":
        print("Slythering")
    case _:
        print("Who?")
```

># Exceptions

###  The "Must Know" Common Exceptions

While there are dozens of exceptions, as a developer, you will encounter these 10 roughly 90% of the time.

| Exception | Cause | Example |
| :--- | :--- | :--- |
| **`SyntaxError`** | You wrote invalid Python code (e.g., missed a colon). | `if x == 5` (missing `:`) |
| **`TypeError`** | Operation on the wrong type of object. | `"Age: " + 25` (cannot add str and int) |
| **`ValueError`** | Right type, but inappropriate value. | `int("hello")` (string is valid type, but content is bad) |
| **`IndexError`** | Trying to access a list index that doesn't exist. | `my_list[10]` (when list has only 3 items) |
| **`KeyError`** | Trying to access a dictionary key that isn't there. | `my_dict["unknown_key"]` |
| **`AttributeError`** | Trying to access a method/variable an object doesn't have. | `variable.append()` (if variable is an Integer) |
| **`NameError`** | Using a variable that hasn't been defined yet. | `print(x)` (where x was never assigned) |
| **`ZeroDivisionError`** | Dividing a number by zero. | `10 / 0` |
| **`FileNotFoundError`** | Trying to open a file that doesn't exist. | `open("missing.txt")` |
| **`ImportError`** | Python cannot find the module you are trying to import. | `import nomodule` |

-----

### The Complete Hierarchy (Python 3.x)

Here is the structured tree of all built-in exceptions.

#### A. The Root Exceptions

These are the fundamental base classes. You usually do **not** catch `BaseException` directly.

  * **`BaseException`**: The root of all exceptions.
      * **`SystemExit`**: Raised by `sys.exit()`.
      * **`KeyboardInterrupt`**: Raised when you press **Ctrl+C**.
      * **`GeneratorExit`**: Raised when a generator is closed.
      * **`Exception`**: **This is the base class for all non-system-exiting exceptions.** (User-defined exceptions should inherit from this).

#### B. Concrete Exceptions (Under `Exception`)

These are the errors you will handle in your logic.

  * **`ArithmeticError`** (Base class for math errors)
      * `FloatingPointError`
      * `OverflowError`
      * `ZeroDivisionError`
  * **`AssertionError`** (Raised by `assert` statement)
  * **`AttributeError`**
  * **`BufferError`**
  * **`EOFError`** (End Of File, e.g., hitting Ctrl+D in `input()`)
  * **`ImportError`**
      * `ModuleNotFoundError`
  * **`LookupError`** (Base for finding items)
      * `IndexError`
      * `KeyError`
  * **`MemoryError`**
  * **`NameError`**
      * `UnboundLocalError`
  * **`OSError`** (Base for Operating System errors)
      * `BlockingIOError`
      * `ChildProcessError`
      * `ConnectionError`
          * `BrokenPipeError`
          * `ConnectionAbortedError`
          * `ConnectionRefusedError`
          * `ConnectionResetError`
      * `FileExistsError`
      * `FileNotFoundError`
      * `InterruptedError`
      * `IsADirectoryError`
      * `NotADirectoryError`
      * `PermissionError`
      * `ProcessLookupError`
      * `TimeoutError`
  * **`ReferenceError`** (Weak reference proxy access)
  * **`RuntimeError`**
      * `NotImplementedError`
      * `RecursionError` (Maximum recursion depth exceeded)
  * **`StopIteration`** (Signals the end of an iterator)
  * **`StopAsyncIteration`**
  * **`SyntaxError`**
      * `IndentationError`
          * `TabError`
  * **`SystemError`** (Internal interpreter error)
  * **`TypeError`**
  * **`ValueError`**
      * `UnicodeError`
          * `UnicodeDecodeError`
          * `UnicodeEncodeError`
          * `UnicodeTranslateError`

#### C. Warnings

These are printed to the console but generally do not stop the program execution unless configured to do so.

  * **`Warning`**
      * `DeprecationWarning` (Feature will be removed in future)
      * `PendingDeprecationWarning`
      * `RuntimeWarning`
      * `SyntaxWarning`
      * `UserWarning`
      * `FutureWarning`
      * `ImportWarning`
      * `UnicodeWarning`
      * `BytesWarning`
      * `ResourceWarning`
