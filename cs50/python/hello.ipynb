># Exceptions

###  The "Must Know" Common Exceptions

While there are dozens of exceptions, as a developer, you will encounter these 10 roughly 90% of the time.

| Exception | Cause | Example |
| :--- | :--- | :--- |
| **`SyntaxError`** | You wrote invalid Python code (e.g., missed a colon). | `if x == 5` (missing `:`) |
| **`TypeError`** | Operation on the wrong type of object. | `"Age: " + 25` (cannot add str and int) |
| **`ValueError`** | Right type, but inappropriate value. | `int("hello")` (string is valid type, but content is bad) |
| **`IndexError`** | Trying to access a list index that doesn't exist. | `my_list[10]` (when list has only 3 items) |
| **`KeyError`** | Trying to access a dictionary key that isn't there. | `my_dict["unknown_key"]` |
| **`AttributeError`** | Trying to access a method/variable an object doesn't have. | `variable.append()` (if variable is an Integer) |
| **`NameError`** | Using a variable that hasn't been defined yet. | `print(x)` (where x was never assigned) |
| **`ZeroDivisionError`** | Dividing a number by zero. | `10 / 0` |
| **`FileNotFoundError`** | Trying to open a file that doesn't exist. | `open("missing.txt")` |
| **`ImportError`** | Python cannot find the module you are trying to import. | `import nomodule` |

-----

### The Complete Hierarchy (Python 3.x)

Here is the structured tree of all built-in exceptions.

#### A. The Root Exceptions

These are the fundamental base classes. You usually do **not** catch `BaseException` directly.

  * **`BaseException`**: The root of all exceptions.
      * **`SystemExit`**: Raised by `sys.exit()`.
      * **`KeyboardInterrupt`**: Raised when you press **Ctrl+C**.
      * **`GeneratorExit`**: Raised when a generator is closed.
      * **`Exception`**: **This is the base class for all non-system-exiting exceptions.** (User-defined exceptions should inherit from this).

#### B. Concrete Exceptions (Under `Exception`)

These are the errors you will handle in your logic.

  * **`ArithmeticError`** (Base class for math errors)
      * `FloatingPointError`
      * `OverflowError`
      * `ZeroDivisionError`
  * **`AssertionError`** (Raised by `assert` statement)
  * **`AttributeError`**
  * **`BufferError`**
  * **`EOFError`** (End Of File, e.g., hitting Ctrl+D in `input()`)
  * **`ImportError`**
      * `ModuleNotFoundError`
  * **`LookupError`** (Base for finding items)
      * `IndexError`
      * `KeyError`
  * **`MemoryError`**
  * **`NameError`**
      * `UnboundLocalError`
  * **`OSError`** (Base for Operating System errors)
      * `BlockingIOError`
      * `ChildProcessError`
      * `ConnectionError`
          * `BrokenPipeError`
          * `ConnectionAbortedError`
          * `ConnectionRefusedError`
          * `ConnectionResetError`
      * `FileExistsError`
      * `FileNotFoundError`
      * `InterruptedError`
      * `IsADirectoryError`
      * `NotADirectoryError`
      * `PermissionError`
      * `ProcessLookupError`
      * `TimeoutError`
  * **`ReferenceError`** (Weak reference proxy access)
  * **`RuntimeError`**
      * `NotImplementedError`
      * `RecursionError` (Maximum recursion depth exceeded)
  * **`StopIteration`** (Signals the end of an iterator)
  * **`StopAsyncIteration`**
  * **`SyntaxError`**
      * `IndentationError`

  * **`TypeError`**

## command-line arguments

eg -> python hello.py

`sys.argv` is a specific variable (a **list**) found in Python's standard `sys` module. It is used to handle **command-line arguments**.

When you run a Python script from your terminal or command prompt, `sys.argv` captures the inputs you provide after the script name and makes them available to your code.

Here is the breakdown of the name:

  * **sys**: The module containing system-specific parameters and functions.
  * **argv**: Short for **"Argument Vector"** (a term inherited from the C programming language), which essentially means a list of argument values.

-----

## üßê How it Works

`sys.argv` is a simple **Python list of strings**.

When you execute a command in your terminal, Python takes that line of text, splits it by spaces, and stores the pieces in this list.

### The Structure of the List

The most important thing to remember is how the indices are assigned:

1.  **`sys.argv[0]`**: This is **always the name of the script** (the program itself).
2.  **`sys.argv[1]`**: This is the **first argument** you passed.
3.  **`sys.argv[2]`**: This is the **second argument**, and so on.

-----

## üíª Code Example

Imagine you have a file named `greet.py` with the following code:

```python
import sys

# Check if the list contains more than just the script name
# (meaning the user actually typed an argument)
if len(sys.argv) > 1:
    # Get the first argument (index 1)
    name = sys.argv[1]
    print(f"Hello, {name}!")
else:
    print("Hello, Stranger!")

# Let's print the whole list to see what it looks like
print(f"\nRaw list data: {sys.argv}")
```

### Running the Code

Here is how this script behaves depending on what you type in your terminal:

**Scenario A: No arguments provided**

```bash
$ python greet.py
```

  * **Output:** `Hello, Stranger!`
  * **Raw list:** `['greet.py']`

**Scenario B: One argument provided**

```bash
$ python greet.py Alice
```

  * **Output:** `Hello, Alice!`
  * **Raw list:** `['greet.py', 'Alice']`

**Scenario C: Multiple arguments provided**

```bash
$ python greet.py Alice Bob
```

  * **Output:** `Hello, Alice!` (The script ignores 'Bob' because we only asked for `sys.argv[1]`)
  * **Raw list:** `['greet.py', 'Alice', 'Bob']`

-----

## ‚ö†Ô∏è Important "Gotchas"

### 1\. Everything is a String

`sys.argv` always stores arguments as **strings**, even if you type numbers. You must convert them manually if you need to do math.

```bash
$ python add.py 5 5
```

If you just do `sys.argv[1] + sys.argv[2]`, Python will output `"55"` (string concatenation).
You must do `int(sys.argv[1]) + int(sys.argv[2])` to get `10`.

### 2\. Spaces Break Arguments

The terminal uses spaces to separate arguments. If you want to pass a single argument that contains spaces (like a full name), you must use quotes.

```bash
$ python greet.py "John Smith"
```

  * `sys.argv[1]` is now `"John Smith"`.
  * Without quotes, `John` would be `[1]` and `Smith` would be `[2]`.
