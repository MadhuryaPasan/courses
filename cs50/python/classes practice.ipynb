```python
class GroceryItem:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def __str__(self):
        return f"{self.quantity}x {self.name} @ ${self.price} each ({self.total_cost()} total)"

    def total_cost(self):
        return self.price * self.quantity


def main():
    item = GroceryItem("Apple", 0.50, 3)
    print(item)


if __name__ == "__main__":
    main()
```

```python
import json
from ollama import chat
from rich import print

# 1. Define the 'Tool' - a simple Python function
def get_weather(city: str) -> str:
    """
    Get the current weather for a specific city.
    Args:
        city: The name of the city (e.g., 'London', 'Tokyo')
    """
    # In a real app, you would call a Weather API here.
    mock_database = {
        "london": {"temp": "15°C", "condition": "Cloudy"},
        "tokyo": {"temp": "22°C", "condition": "Sunny"},
        "new york": {"temp": "10°C", "condition": "Rainy"}
    }
    data = mock_database.get(city.lower(), {"temp": "unknown", "condition": "unknown"})
    return json.dumps(data)

def run_agent_session():
    model_name = 'functiongemma:270m' # You can also use 'functiongemma'
    
    # The conversation history
    messages = [{'role': 'user', 'content': 'What is the weather like in Tokyo?'}]
    
    print(f"[bold cyan]User:[/bold cyan] {messages[0]['content']}")

    # 2. First Call: Let Gemma 3 decide to use the tool
    response = chat(
        model=model_name,
        messages=messages,
        tools=[get_weather], # We pass the function directly as a tool
    )

    # 3. Process Tool Calls
    if response.message.tool_calls:
        for tool in response.message.tool_calls:
            print(f"[bold yellow]Agent Action:[/bold yellow] Calling {tool.function.name} with {tool.function.arguments}")
            
            # Execute the actual Python function
            result = get_weather(**tool.function.arguments)
            
            # Add the model's intent and the tool's result to the history
            messages.append(response.message)
            messages.append({'role': 'tool', 'content': result})

        # 4. Final Call: Let Gemma 3 summarize the result for the user
        final_response = chat(model=model_name, messages=messages)
        print(f"[bold green]Gemma 3:[/bold green] {final_response.message.content}")
    else:
        print(f"[bold green]Gemma 3:[/bold green] {response.message.content}")

if __name__ == "__main__":
    run_agent_session()
```
