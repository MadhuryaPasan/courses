```python
class GroceryItem:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def __str__(self):
        return f"{self.quantity}x {self.name} @ ${self.price} each ({self.total_cost()} total)"

    def total_cost(self):
        return self.price * self.quantity


def main():
    item = GroceryItem("Apple", 0.50, 3)
    print(item)


if __name__ == "__main__":
    main()
```
