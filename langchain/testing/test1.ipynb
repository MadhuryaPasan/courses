```python
import os

# MODEL = "qwen3:0.6b"
# MODEL = "qwen3:1.7b"
MODEL = "qwen3:4b"
BASE_URL = "http://localhost:11434/v1"

os.environ["OPENAI_API_KEY"] = "none"
```

```python
from IPython.display import display, Markdown
```

## Using Ollama to display AI response with the reasoning

```python
from langchain_ollama import ChatOllama

# Initialize the chat model
ollama_llm = ChatOllama(model=MODEL, reasoning=True)

response = ollama_llm.invoke("hello")

display(Markdown(response.additional_kwargs["reasoning_content"]))

print("==========================================")

display(Markdown(response.content))
```

```python
# a way to properly display the output
from pprint import pprint


pprint(response.content)
pprint(response.additional_kwargs)
pprint(response.response_metadata)
pprint(response.id)
```

## Using OpenAI to display AI response with the reasoning

```python
from langchain_openai import ChatOpenAI

reasoning = {
    "effort": "low",  # 'low', 'medium', or 'high'
    "summary": "auto",  # 'detailed', 'auto', or None
}

openai_llm = ChatOpenAI(
    model=MODEL, 
    base_url=BASE_URL, 
    reasoning=reasoning
    )

response = openai_llm.invoke("Hello")

display(Markdown(response.content[0]["summary"][0]["text"]))

print("==========================================")

display(Markdown(response.content[1]["text"]))
```

```python
pprint(response.content)
pprint(response.response_metadata)
```
